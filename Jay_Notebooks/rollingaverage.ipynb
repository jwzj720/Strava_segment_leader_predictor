```python
import matplotlib.pyplot as p
import random as r
```

```python
def randomset(n):
    a = []
    for i in range(n):
        a.append(r.randint(0,100))
    return a

def roll_avg(old_arr,scale):
    new_arr = []
    current = []
    
    #iterate through original array
    for i in range(len(old_arr)):
        #add current value to array for averaging
        current.append(old_arr[i])
        #if enough values are present (or end has been reached) average values in current and add to new_arr
        
        #if i==len(old_arr)-1:
        #    print("End segment: "+str(scale)+" - "+str(len(current)))
            
            
        if len(current)==scale or i==len(old_arr)-1:
            new_arr.append(sum(current) / len(current))
            current = []
    
    return new_arr
            
#repeatedly applies roll_avg to an array, option to show for each iteration
def roll_avg_loop(old_arr,scale,iterations,showall):
    new_arr = roll_avg(old_arr,scale)
    if showall == True or iterations==0:
        p.plot(new_arr)
        p.show()
    if iterations>0:
        roll_avg_loop(new_arr,scale,iterations-1,showall)
    
```

```python
#tests averaging a basic set by different increments
set_a = randomset(1000)
set_b = roll_avg(set_a,2)
set_c = roll_avg(set_a,5)
set_d = roll_avg(set_a,10)
set_e = roll_avg(set_a,50)


p.plot(set_a)
p.show()

p.plot(set_b)
p.show()

p.plot(set_c)
p.show()

p.plot(set_d)
p.show()

p.plot(set_e)
p.show()
```

```python
#test averaging the same set repeatedly, small chunks averaged 
set_a = randomset(10000)
roll_avg_loop(set_a,2,8,True)
```

```python
#test averaging the same set repeatedly, larger chunks averaged fewer times
set_a = randomset(10000)
roll_avg_loop(set_a,5,4,True)
```

```python
#test averaging the same set repeatedly, even larger chunks averaged even fewer times
set_a = randomset(10000)
roll_avg_loop(set_a,10,2,True)
```
